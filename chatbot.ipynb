```python
def chatbot():
    print("🤖 Chatbot: Hello! Type 'bye' to exit.")

    while True:
        user_input = input("You: ").lower()

        if user_input == "bye":
            print("🤖 Chatbot: If You Want Anything,You can ask me directly.")
            print("🤖 Chatbot: Goodbye! Have a nice day 😊")
            break

        elif user_input in ["hi", "hello", "hey"]:
            print("🤖 Chatbot: Hello! How can I help you?")

        elif "how are you" in user_input:
            print("🤖 Chatbot: I'm just a bot, but I'm doing great! 😊")

        elif "your name" in user_input:
            print("🤖 Chatbot: CHATBOT! You can give me any Nickname.")

        elif "what can you do" in user_input:
            print("🤖 Chatbot: I can respond to greetings and simple questions.")
        elif "Had Breakfast" in user_input:
            print("🤖 Chatbot: I'm a machine,so i can't ate.")

        else:
            print("🤖 Chatbot: Sorry, I don't understand that.")

# Run chatbot
chatbot()
```
